```python
import pandas as pd
import numpy as np
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies.head()
```

```python
credits=credits.rename(columns={'movie_id':'id'})
```

```python
credits.head()
```

```python
movies=movies.merge(credits, on='id')
```

```python
movies.head()
```

```python
movies.describe()
```

```python
movies.drop(columns=['homepage','status','crew','spoken_languages','production_countries','release_date','title_x','title_y'])
```

```python
R=movies['vote_average']
v=movies['vote_count']
C=movies['vote_average'].mean()
m=movies['vote_count'].quantile(0.80)
```

```python
movies['Weighted Average']=((R*v)+(C*m))/(v+m)
```

```python
movies.head()
```

```python
movies=movies.sort_values('Weighted Average', ascending=False)
```
